# Prototype data processing
A notebook to make the data processing and testing to be written in the proper organized way in the github folder.

# Import 

```python
import pandas as pd 
import numpy as np
import sys 
```

```python
df = pd.read_csv('../data/raw_data/ds.csv')
```

# Clean/organize ds 

```python
df.head()
```

```python
col_to_save = [
    'a1_mean', 'a2_mean', 'd1_mean', 'd2_mean', 'd3_mean', 'd4_mean', 'd5_mean', 'id '
]

df = df.drop(columns = [col for col in df.columns if col not in col_to_save])

df.head()
```

```python
raname_dict = {
    'a1_mean': 'aggregated1',
    'a2_mean': 'aggregated2',
    'd1_mean': 'granular1',
    'd2_mean': 'granular2',
    'd3_mean': 'granular3',
    'd4_mean': 'granular4',
    'd5_mean': 'granular5',
    'id ': 'id'
}

df = df.rename(columns = raname_dict)

df.head()
```

```python
df.to_csv('../data/interim_data/clean_ds.csv', index = False)
```

# Process the dataset for the model

```python
mean_dict = {}
std_dict = {}
for col in list(df.columns):
    mean_dict[col] = df[col].mean()
    std_dict[col] = df[col].std()
```

```python
for col in list(set(df.columns) - set(['id'])):
    df[col] = (df[col] - mean_dict[col]) / std_dict[col]
```

```python
df.to_csv('../data/processed_data/normalized_ds.csv')
```
